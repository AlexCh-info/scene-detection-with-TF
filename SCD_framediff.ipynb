```python
import cv2
import numpy as np
```

```python
def gray_scale_preproccessing(frame):
  return cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
```

```python
def blur_preproccessing(frame):
  return cv2.GaussianBlur(frame, (5, 5), 0)
```

```python
def resize_preproccessing(frame, resize=(128, 128)):
  return cv2.resize(frame, resize)
```

```python
def preproccessing_frame(frame, resize=(128,128)):
  gray = cv2.cvtColor(frame, cv2.COLOR__BGR2GRAY)
  gray = cv2.GaussianBlur(gray, (5,5), 0)
  gray = cv2.resize(gray, resize)
  return gray
```

```python
def frame_diff_changed_ratio(frame1, frame2, thresh_val=25, resize=None):
  if frame1.ndim == 3:
    f1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
    f2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
  else:
    f1, f2 = frame1, frame2

  if resize is not None:
    f1 = cv2.resize(f1, resize)
    f2 = cv2.resize(f2, resize)

  diff = cv2.absdiff(f1, f2)

  _, thresh = cv2.threshold(diff, thresh_val, 255, cv2.THRESH_BINARY)
  return np.count_nonzero(thresh) / thresh.size
```

```python
def frame_diff(frame1, frame2):
  frame1 = gray_scale_preproccessing(frame1)
  frame2 = gray_scale_preproccessing(frame2)

  frame1 = blur_preproccessing(frame1)
  frame2 = blur_preproccessing(frame2)

  rz_frame1 = resize_preproccessing(frame1, (128, 128))
  rz_frame2 = resize_preproccessing(frame2, (128,128))

  diff = cv2.absdiff(rz_frame1, rz_frame2)
  return np.mean(diff)
```

```python
def detect_scene(video_path, threshold = 2.0, thresh_val=25, resize=(128,128)):
  cap = cv2.VideoCapture(video_path)
  prive_frame = None
  diffs = []


  while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
      break
    if prive_frame is not None:
      d = frame_diff_changed_ratio(prive_frame, frame, thresh_val=thresh_val, resize=resize)
      diffs.append(d)
    prive_frame = frame.copy()
  cap.release()

  diffs = np.array(diffs)
  mean = np.mean(diffs)
  std = np.std(diffs)
  if std == 0:
    threshold_new = mean + 0.1
  else:
    threshold_new = mean + threshold * std

  scene_change = []
  for i, d in enumerate(diffs):
    if d > threshold_new:
      scene_change.append(i + 1)
  return scene_change, diffs
```

```python
def seconds(frame, video_path):
  cap = cv2.VideoCapture(video_path)
  fps = cap.get(cv2.CAP_PROP_FPS)
  second = frame / fps
  return second, fps
```
